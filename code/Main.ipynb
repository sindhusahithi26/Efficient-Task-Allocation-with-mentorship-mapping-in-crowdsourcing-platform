```python
import pandas as pd

# Read the updated dataset
data = pd.read_csv('Updated Employee Dataset.csv')

# Filter for inexpert employees who need mentors
inexpert_data = data[data['Employee Type'] == 'Inexpert']

# Define a function to find an expert mentor for an inexpert employee
def assign_mentor(row):
    mentors = data[(data['Employee Type'] == 'Expert') & (data['Skill Set'] == row['Skill Set']) & (data['SuccessFull Completions Percentage'] >= 40)]
    
    if not mentors.empty:
        mentor = mentors.iloc[0]
        return pd.Series({'Mentor ID': mentor['Emp.id'], 'Mentor Name': mentor['Employee Name']})
    else:
        return pd.Series({'Mentor ID': None, 'Mentor Name': None})

# Apply the mentor assignment function to each inexpert employee
mentors_df = inexpert_data.apply(assign_mentor, axis=1)

# Combine the mentor information with the inexpert employee data
inexpert_data = pd.concat([inexpert_data, mentors_df], axis=1)

# Save the updated dataset with mentor assignments
inexpert_data.to_csv('Inexpert Employee Dataset with Mentors.csv', index=False)

print("Mentor assignment completed and saved to 'Inexpert Employee Dataset with Mentors.csv'.")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load the employee dataset from a CSV file
df = pd.read_csv("Employee Dataset.csv")

# Categorize employees based on experience
df['Category'] = df['Experience'].apply(lambda x: 'Expert' if x > 4 else 'Inexpert')

# Create a mapping for expert employees who are eligible for mentorship
expert_eligible_mapping = df[(df['Category'] == 'Expert') &
                             (df['SuccessFull Completions Percentage'] >= 50) &
                             (df['Approved OnTime'] >= 40)]

# Ensure that the Skill Set column is treated as a string
df['Skill Set'] = df['Skill Set'].astype(str)

# Separate experts and inexperts
experts = df[df['Category'] == 'Expert']
inexperts = df[df['Category'] == 'Inexpert']
print(inexperts.count())
print(experts.count())
print(expert_eligible_mapping.count())

# Vectorize the skill sets using TF-IDF for both experts and inexperts
vectorizer = TfidfVectorizer()
tfidf_matrix_experts = vectorizer.fit_transform(experts['Skill Set'])
tfidf_matrix_inexperts = vectorizer.transform(inexperts['Skill Set'])

# Calculate the cosine similarity between inexperts and eligible expert mentors
cosine_similarities = cosine_similarity(tfidf_matrix_inexperts, tfidf_matrix_experts)

# Define the maximum number of inexperts to be mapped to a single expert mentor
max_inexperts_per_mentor = 2
mentor_count = {}

# Find the most similar eligible expert for each inexpert
inexpert_to_mentor_mapping = pd.DataFrame(columns=['Inexpert Employee', 'Expert Mentor', 'Similarity Score'])

for i, row in enumerate(cosine_similarities):
    # Find the index of the most similar eligible expert
    eligible_experts_indices = expert_eligible_mapping.index
    sorted_indices = np.argsort(row)[::-1]

    # Map up to max_inexperts_per_mentor inexperts to the expert mentor
    mapped_inexperts = 0
    for idx in sorted_indices:
        if idx in eligible_experts_indices:
            # Check if the mentor can accept more inexperts
            if mentor_count.get(idx, 0) < max_inexperts_per_mentor:
                # Check if the inexpert has not been mapped before
                if inexperts['Emp.id'].iloc[i] not in inexpert_to_mentor_mapping['Inexpert Employee'].values:
                    # Update the mentor_count dictionary
                    mentor_count[idx] = mentor_count.get(idx, 0) + 1

                    # Get the similarity score
                    similarity_score = row[idx]

                    # Print the original inexpert employee and the mapped expert mentor
                    print(f"Inexpert Employee {inexperts['Emp.id'].iloc[i]} is mapped to Expert Mentor {df['Emp.id'].iloc[idx]} with Similarity Score: {similarity_score}")

                    # Add the mapping to the DataFrame
                    inexpert_to_mentor_mapping = inexpert_to_mentor_mapping.append(
                        {'Inexpert Employee': inexperts['Emp.id'].iloc[i],
                         'Expert Mentor': df['Emp.id'].iloc[idx],
                         'Similarity Score': similarity_score}, ignore_index=True)

                    mapped_inexperts += 1

        if mapped_inexperts >= max_inexperts_per_mentor:
            break  # Break after mapping to the allowed number of expert mentors

# Print the resulting mappings
print("Inexpert to Expert Mentor Mapping:")
print(inexpert_to_mentor_mapping)

# Save the mappings to a CSV file
inexpert_to_mentor_mapping.to_csv("inexpert_to_mentor_mapping.csv", index=False)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Read the Employee and Tasks datasets
employee_data = pd.read_csv("Employee dataset.csv")
tasks_data = pd.read_csv("Tasks Dataset.csv")

# Define the mapping function to categorize employees
def map_employee_to_category(row):
    experience = row['Experience']
    success = row['SuccessFull Completions Percentage']
    
    if experience > 4 and success > 40:
        return 'Expert'
    else:
        return 'Inexpert'

# Apply the mapping function to create a new 'Category' column
employee_data['Category'] = employee_data.apply(map_employee_to_category, axis=1)
num_experts = employee_data[employee_data['Category'] == 'Expert'].shape[0]
num_inexperts = employee_data[employee_data['Category'] == 'Inexpert'].shape[0]

# Print the counts
print(f"Number of Experts: {num_experts}")
print(f"Number of Inexperts: {num_inexperts}")

# Calculate skills similarity using cosine similarity
vectorizer = CountVectorizer()
employee_skills_matrix = vectorizer.fit_transform(employee_data['Skill Set'].apply(lambda x: ' '.join(str(x).split(','))))
task_skills_matrix = vectorizer.transform(tasks_data['Task Requirement Skills'].apply(lambda x: ' '.join(str(x).split(','))))

skills_similarity_matrix = cosine_similarity(employee_skills_matrix, task_skills_matrix)

# Track assignments for each expert and inexpert
task_assignment = {}

# Track assigned tasks count for each employee
assigned_tasks_count = {employee: 0 for employee in employee_data['Employee Name']}

# Iterate through tasks and allocate to employees
for task_index, task_row in tasks_data.iterrows():
    task_id = task_row['Task id']
    if task_id not in task_assignment:
        # Find the index of the employee with the highest skills similarity for the task
        employee_index = skills_similarity_matrix[:, task_index].argmax()
        similarity_level = skills_similarity_matrix[employee_index, task_index]

        # Assign task based on level of difficulty and due days criteria
        level_of_difficulty = task_row['Level of Urgency']
        due_days = task_row['Due Date in days']

        # Check if the employee is available and the task is not assigned to the employee
        if (
            (level_of_difficulty == 'Low' and employee_data.at[employee_index, 'Category'] == 'Inexpert' and assigned_tasks_count[employee_data.at[employee_index, 'Employee Name']] < 2) or
            (level_of_difficulty == 'High' and employee_data.at[employee_index, 'Category'] == 'Expert' and assigned_tasks_count[employee_data.at[employee_index, 'Employee Name']] < 3) or  
            (level_of_difficulty == 'Medium' and due_days < 25 and employee_data.at[employee_index, 'Category'] == 'Inexpert' and assigned_tasks_count[employee_data.at[employee_index, 'Employee Name']] < 2) or
            (level_of_difficulty == 'Medium' and (due_days >= 25 or employee_data.at[employee_index, 'Category'] == 'Expert') and assigned_tasks_count[employee_data.at[employee_index, 'Employee Name']] < 3)  # Adjusted due date constraint
        ):
            task_assignment[task_id] = (employee_data.at[employee_index, 'Employee Name'], similarity_level)
            assigned_tasks_count[employee_data.at[employee_index, 'Employee Name']] += 1

# Print the resulting task allocation with similarity level
for task_id, (assigned_employee, similarity_level) in task_assignment.items():
    print(f"Task ID: {task_id} -> Assigned to: {assigned_employee} with Similarity Level: {similarity_level:.4f}")

print(f"Total tasks assigned: {len(task_assignment)}")

# Create a DataFrame for the results
result_df = pd.DataFrame(columns=['Task ID', 'Assigned Employee', 'Similarity Level'])

# Populate the DataFrame with assignments and similarity levels
for task_id, (assigned_employee, similarity_level) in task_assignment.items():
    result_df = pd.concat([result_df, pd.DataFrame({'Task ID': [task_id], 'Assigned Employee': [assigned_employee], 'Similarity Level': [similarity_level]})])

# Print the resulting task allocation with similarity level
print(result_df)

# Save the DataFrame to a CSV file
result_df.to_csv('task_assignments3.csv', index=False)
```
